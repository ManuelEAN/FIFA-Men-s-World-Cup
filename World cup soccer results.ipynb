# Soccer World Cup results 1930-2018

In the following dataset, I'm going to evaluate and extract some interesting datas about the results of the Soccer World Cup.
For this excersice, I'm going to be using libraries such as Pandas, Numpy, Seaborn and Matplotlib.


```python
import pandas as pd 
import re
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
df = pd.read_csv('matches_1930_2018.csv')
pd.set_option("display.max_columns", 25)
df.head()
```

#  1.a) Games that played extra time

```python
#Matches that requiered extra time
df[(df['Notes'] == 'Required Extra Time')]
```

# 1.b) Countries with more games in extra time

```python
df_extra = df[(df['Notes'] == 'Required Extra Time')]
```

```python
df_extra_teams = df_extra[['home_team', 'away_team']]
```

```python
df_countries = df_extra_teams.stack().value_counts()
df_extra_teams.stack().value_counts()
```

```python
df_countries.plot.bar(xlabel='Teams' , ylabel='# of Matches', legend=None )
plt.title("Teams with more OT Matches")
plt.show()
```

# 2.) Top 5 games with higher attendance

```python
df.sort_values('Attendance', ascending = False).head(5)
```

# 3.) Analysis about games after group stage 


```python
df_playoff = df[(df['Round'] == 'Round of 16') | (df['Round'] == 'Quarter-finals') | (df['Round'] == 'Semi-finals') | (df['Round'] == 'Final')]
```

# 3.a) Top 5 differences in matches in playoff rounds

```python
df_playoff['Goal Difference'] = pd.Series.abs(df['home_score'] - df['away_score'])
df_playoff.sort_values('Goal Difference' , ascending = False).head(5)
```

# 3.b) Countries with more games in playoff rounds

```python
df_playoff_teams = df_playoff[['home_team', 'away_team']]
df_playoff_teams.stack().value_counts().head(10)
```

# 4.a) Top 5 World Cup Events with more attendace

```python
df_att = df.groupby('Year').agg(Overall_attendance = ('Attendance', 'sum'))

df_att.sort_values(by=['Overall_attendance'], ascending=False).head()
```

# 4.b) Graphic of attendance throughout the years

```python
df_att.plot(xlabel='Year' , ylabel='Attendance(Million)', legend=None, color='magenta' )
plt.title('Soccer World Cup Attendance history')
plt.show()
```

# 5.a) Teams with more Championships

```python
df_finals = df[(df['Round'] == 'Final')]

#Converting penalty results to INT64
df_finals['home_penalty'] = df['home_penalty'].astype('Int64')
df_finals['away_penalty'] = df['away_penalty'].astype('Int64')
```

```python
#Defining Scores (Regular time + OT + Penalties) as int64
df_finals['Total_Home'] = (df_finals.fillna(0)['home_score'] + df_finals.fillna(0)['home_penalty']).astype('Int64')
df_finals['Total_Away'] = (df_finals.fillna(0)['away_score'] + df_finals.fillna(0)['away_penalty']).astype('Int64')
```

```python
#Definning column Winner of the Total Result
df_finals.loc[df_finals['Total_Home'] - df_finals['Total_Away'] > 0, 'Winner'] = (df_finals['home_team'])
df_finals.loc[df_finals['Total_Home'] - df_finals['Total_Away'] < 0, 'Winner'] = (df_finals['away_team'])
```

```python
#Changing West Germany to Germany due to change in FIFA name
df_finals['Winner'].mask(df_finals['Winner'] == 'West Germany', 'Germany', inplace=True)
```

```python
df_finals['Winner'].value_counts()
```

```python
df_finals['Winner'].value_counts().plot(kind='bar', color = 'yellow')
plt.xlabel('Country')
plt.ylabel('# Soccer World Cup Championships')
plt.title("Soccer World Cup Championships by Teams")
plt.show()
```
